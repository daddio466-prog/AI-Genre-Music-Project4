### **Question 1**
### Do longer AI music prompts lead to higher user ratings?



```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure ratings DataFrame is available for plotting
ratings = pd.read_csv("AI-Genre-Music-Project4/ai_generated_song_ratings.csv")

plt.figure(figsize=(8,6))
plt.scatter(ratings['prompt_length'], ratings["user_rating"])
plt.title("Prompt Length vs. User Rating")
plt.xlabel("Prompt Length")
plt.ylabel("User Rating")
plt.tight_layout()
plt.show()
```

The scatter plot displays the relationship between prompt length and user ratings. From the visualization there appears to be no significant correlation between user ratings and prompt length.

### Which genres are used more in AI-generated music prompts?

```python
genre_counts = prompts["genre"].value_counts()

plt.figure(figsize=(8,6))
genre_counts.plot(kind="bar")
plt.title("Frequency of Genres in AI Music Prompts")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The genres used most in AI-generated music prompts are orchestral, lo-fi, and rock.

```python
from pandas.plotting import scatter_matrix

scatter_matrix(ratings[["prompt_length", "user_rating", "listen_count"]], figsize=(8,8))
plt.suptitle("Scatter Matrix of Prompt Length, User Rating, and Listen Count")
plt.show()
```

Diagonal plots (histograms): These show the distribution of each individual variable. For example, you can see the distribution of prompt lengths, user ratings, and listen counts.
Off-diagonal plots (scatter plots): These illustrate the pairwise relationships between the variables. For instance, you can observe the relationship between prompt_length and user_rating (reiterating our previous finding of no strong correlation), prompt_length and listen_count, and user_rating and listen_count.

```python
corr = ratings[["prompt_length", "user_rating", "listen_count"]].corr()

plt.figure(figsize=(8,8))
plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix of Prompt Length, User Rating, and Listen Count")
plt.tight_layout()
plt.show()
```

From the correlation matrix, the correlation coefficient between 'user_rating' and 'listen_count' is approximately 0.027. This is a very low positive value, indicating that there is almost no linear relationship between how highly a song is rated by users and how many times it is listened to. In simpler terms, a higher user rating does not significantly predict a higher listen count, and vice versa, based on this dataset.
